```python
import json
import pandas as pd

# increase the number of columns that appear when print pandas dataframe
pd.set_option("display.max_columns", 50)
```

Start by loading the 'tweet_json.txt' file as a pandas dataframe. The code for gathering data using Tweepy api is in the gather_data.ipynb file included in the submission folder. 

```python
# define the file name for tweet_json.txt
tweet_file = 'tweet_json.txt'


# open the file for reading
with open(tweet_file, 'r') as f:
    
    # read all rows
    rows = f.readlines()
    
    
    for i, row in enumerate(rows):
        
        # load one line and convert it from json format to dict
        row_dict = json.loads(row)
        
        # create an empty dataframe with column names
        if i == 0:
            tweet_details_df = pd.DataFrame(columns = row_dict.keys())
        
        # append the dictionary to the created dataframe
        tweet_details_df = tweet_details_df.append(row_dict, ignore_index= True )
```

### Define

- Melt the `doggy`, `floofer`, `puppy` and `puppo` in the `tweet_arch`  into a `dog stage` column.

### Code

Before proceeding to coding this task, let's check whether there are any duplicates in the `tweet_id` column. I am doing this because after the melting task I am performing, if there are more than one dog stage associated to a given `tweet_id`, the melting operation will produce duplicates in the `tweet_id` column. This would indicate a one-to-many relationship into tweets and dog stages and might required dog stages to be separated into a new dataframes

```python
# check for duplicate ids

tweet_arch_clean_df.tweet_id[tweet_arch_clean_df.tweet_id.duplicated()]
```

Nice, no duplicates are initially present

Now we proceed to the melting operation

```python
# Melt the `doggy`, `floofer`, `puppy` and `puppo` in the `tweet_arch`  into a `dog stage` column.

dog_stages_df = pd.melt(frame = tweet_arch_clean_df, 
                       id_vars = ['tweet_id', 'timestamp', 'text', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp', 'expanded_urls', 'rating_numerator', 'rating_denominator', 'name'],  
                       var_name = 'type_name', 
                       value_name = 'dog_stage')

```

```python
# Remove rows were the dog stage is None
dog_stages_df = dog_stages_df[dog_stages_df.dog_stage != "None"]
```

```python

# drop the 'type_name' column since it adds no information
dog_stages_df = dog_stages_df.drop(columns= [ 'type_name'])
```

check out whether the melting operation is successful

```python
dog_stages_df.sample(10)
```

Awesome! the melting operation is successful. A `dog_stage` column is created.



Let's now check out whether we have duplicate `tweet_id` values with multiple dog stages

```python
# get a table of with duplicate `tweet_id` values
dup_columns = dog_stages_df[dog_stages_df.tweet_id.duplicated(keep = False)]

# sort the rows by `tweet_id` to show rows with duplicates next to each other
dup_columns = dup_columns.sort_values(by = ['tweet_id'])

# show the result
dup_columns[['tweet_id', 'dog_stage']]
```

We clearly notice that there are some duplicates. This indicates that there is a one-to-many relationship between tweet ids and dog stages. The new dataframe called 'dog_stages_df' should be kept. It should only contain tweet ids and dog stages. The `doggo`, `floofer`, `pupper` and `puppo` columns should be removed from the `tweet_arch_clean` dataframe.

```python
# keep only the `tweet_id` and the `dog_stage` columns in the `dog_stages` dataframe
dog_stages_df = dog_stages_df[['tweet_id', 'dog_stage']]

# Remove the `doggo`, `floofer`, `pupper` and `puppo` columns from the `tweet_arch_clean` dataframe.
tweet_arch_clean_df = tweet_arch_clean_df.drop(columns = ['floofer', 'pupper', 'doggo', 'puppo'])
```

# Analysis and Insights

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Q1: what is the distribution of rating numerators for those with a denominator of 10?

Let's first check for the range of values of the numerator between the 1% and the 99% quantiles:

```python
tweet_arch_master_df.rating_numerator.describe(percentiles=[0.01, 0.99])
```

```python
# let's take the quantile values
q1 = tweet_arch_master_df.rating_numerator.quantile(q = 0.01)
q99 = tweet_arch_master_df.rating_numerator.quantile(q = 0.99)
(q1, q99)
```

```python
# Now let's take only the values of the numerator in this quantile range
rating_nums = tweet_arch_master_df.query('rating_numerator >= @q1 & rating_numerator <= @q99').rating_numerator
```

```python
# and finally plot the distribution of rating numerators in that range
sns.barplot(x = rating_nums.unique(), y = rating_nums.value_counts())
plt.xlabel('rating numerator')
plt.ylabel('count')
plt.title('The distribution of rating numerators')
plt.show()
```

My observation is that most dogs are rated more than the denominator of 10. This is characteristic of the WeRateDogs dataset. We also notice that although rating is exagerated in this way, highly exagerated ratings (> 14/10) are quite rare. Low rating (< 10/10) are also rare except for a peak at 5/10 rating.

### Q2: what are the highest rated dog stages?

To answer this question lets plot the average rating numerator per dog stage.

```python
# Let's first join the 'dog_stages' dataframe with the 'tweet_arch_master' dataframe
dog_ratings_df = dog_stages_df.merge(tweet_arch_master_df, how = 'left', on = 'tweet_id')[['rating_numerator', 'dog_stage']]
```

```python
# Take only those rows corresponding to rating numerators between the 1% and 99% quantiles as in the previous questions
dog_ratings_df = dog_ratings_df.query('rating_numerator >= @q1 & rating_numerator <= @q99')
```

```python
# create a boxplot to answer the above question
sns.boxplot(x = 'dog_stage', y = 'rating_numerator', data = dog_ratings_df)
plt.xlabel('Dog stage')
plt.ylabel('Rating numerator (denominator is 10)')
plt.title('Rating numerators per dog stage')
plt.show()
```

Oh, it seems that dog stages are highly imbalanced. So in order to make a decision  of which dog stage is most rated I would suggest the need for hypothesis testing.
